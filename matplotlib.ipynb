```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
```

# basic graph

```python
x = [0,1,2,3,4]
y=[0,2,4,6,8]
plt.plot(x,y)
#adding title to graph
plt.title("this is a graph",fontdict={'fontname':'Comic Sans MS','fontsize':20})

#adding labels 
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

#to remove labels
# plt.legend()
#show graph
plt.show()
```

```python
plt.plot(x,y,label='2x',color='red',linewidth=2,marker='.',linestyle='--',markersize=10,markeredgecolor='blue')
#to remove decimal points 
plt.xticks([0,1,2,3])
plt.yticks([0,2,4,6,8,10])

plt.show()
```

```python
##line number two 
x2 = np.arange(0,4.5,0.5)
print(x2)


#resize graph
plt.figure(figsize=(5,3),dpi=200)


# plt.plot(x2,x2**2,label='x^2')

#plot part of the graph as line
plt.plot(x2[:6],x2[:6]**2,'r',label='x^2')
#plot part of the graph as dot
plt.plot(x2[5:],x2[5:]**2,'r--',label='x^2')


#save fig
plt.savefig('mygraph',dpi=300)
```

```python
#BAR CHART

labels=['A','B','C']
values = [1,4,2]
bars=plt.bar(labels,values)

patterns=['/','*','O']

for i in bars:
    i.set_hatch(patterns.pop(0))

# bars[0].set_hatch('/')
# bars[1].set_hatch('-')
# bars[2].set_hatch('O')



plt.figure(figsize=(6,4))

plt.show()
```
