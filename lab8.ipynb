```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load the Boston dataset
boston = fetch_openml(name='boston', version=1, as_frame=True)
X = boston.data
y = boston.target
feature_names = boston.feature_names

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Initialize PCA for 4 components
pca_4 = PCA(n_components=4)
X_transformed_4 = pca_4.fit_transform(X_scaled)

# Initialize PCA for 3 components
pca_3 = PCA(n_components=3)
X_transformed_3 = pca_3.fit_transform(X_scaled)

# Initialize PCA for 2 components
pca_2 = PCA(n_components=2)
X_transformed_2 = pca_2.fit_transform(X_scaled)

# Plotting
fig, axes = plt.subplots(3, 1, figsize=(10, 15))

# PCA with 4 components
axes[0].scatter(X_transformed_4[:, 0], X_transformed_4[:, 1], c=y, cmap='viridis', edgecolor='k')
axes[0].set_title('PCA with 4 Components')
axes[0].set_xlabel('Principal Component 1')
axes[0].set_ylabel('Principal Component 2')
axes[0].grid(True)

# PCA with 3 components
axes[1].scatter(X_transformed_3[:, 0], X_transformed_3[:, 1], c=y, cmap='viridis', edgecolor='k')
axes[1].set_title('PCA with 3 Components')
axes[1].set_xlabel('Principal Component 1')
axes[1].set_ylabel('Principal Component 2')
axes[1].grid(True)

# PCA with 2 components
axes[2].scatter(X_transformed_2[:, 0], X_transformed_2[:, 1], c=y, cmap='viridis', edgecolor='k')
axes[2].set_title('PCA with 2 Components')
axes[2].set_xlabel('Principal Component 1')
axes[2].set_ylabel('Principal Component 2')
axes[2].grid(True)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from mglearn.datasets import load_extended_boston
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import seaborn as sns  # Import seaborn for visualizations

# Load the Extended Boston Housing Dataset
X, y = load_extended_boston()

# Standardize the Data (PCA works better with standardized data)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Fit PCA Models for Different Dimensions
pca_4 = PCA(n_components=4)
pca_4.fit(X_scaled)

pca_3 = PCA(n_components=3)
pca_3.fit(X_scaled)

pca_2 = PCA(n_components=2)
pca_2.fit(X_scaled)

# Transform the Data
X_transformed_4 = pca_4.transform(X_scaled)
X_transformed_3 = pca_3.transform(X_scaled)
X_transformed_2 = pca_2.transform(X_scaled)

# Visualize the Transformation Results with Seaborn

# Visualize for 2 dimensions
sns.scatterplot(
    x=X_transformed_2[:, 0],
    y=X_transformed_2[:, 1],
    hue=y,  # Color points based on target variable (house price)
    palette="viridis",  # Use a color palette for better visualization
)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("Extended Boston Housing Data Transformed to 2 Dimensions using PCA")
plt.show()

# Visualize for 3 dimensions (using a pair plot)
sns.pairplot(pd.DataFrame(X_transformed_3, columns=["PC1", "PC2", "PC3"]))  # Create a dataframe with principal components
plt.suptitle("Extended Boston Housing Data Transformed to 3 Dimensions using PCA")
plt.show()

# VISUALISE FOR 4 DIMENSIONS USING EXPLAINED VARIANCE.
print("Explained Variance Ratio for 4 dimensions:", pca_4.explained_variance_ratio_)





```
